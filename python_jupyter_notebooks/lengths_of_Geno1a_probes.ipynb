```python
from Bio import SeqIO


def analyze_fasta_lengths(fasta_file):
    lengths = []

    # Read sequences from the FASTA file
    for record in SeqIO.parse(fasta_file, "fasta"):
        lengths.append(len(record.seq))
        print(f"Sequence: {record.id}, Length: {len(record.seq)}")

    if lengths:
        shortest = min(lengths)
        longest = max(lengths)
        print(f"\nShortest FASTA length: {shortest}")
        print(f"Longest FASTA length: {longest}")
        print(
            f"\nThe lengths of the FASTA sequences range from {shortest} to {longest} base pairs."
        )
    else:
        print("No sequences found in the file.")


# Specify the path to your FASTA file
probes = "../msa/blast_to_control/Geno1a_probes.fasta"

# Run the analysis
analyze_fasta_lengths(probes)
```
